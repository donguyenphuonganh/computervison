```python
# c
def dequy(n):
  if n==0: return 1
  if n==1: return 1
  return dequy(n-1)*n
num = int(input("nhap n: "))
print(dequy(num))
```

```python
# d
dict = {}
def dic(n):
  mdict = {}
  for i in range (1,n+1):
    mdict[i] = i*i
  return mdict

n = int(input('nhap n: '))
dict = dic(n)
dict
```

```python
#f
from datetime import datetime
class employee:
    def __init__(self):
        self.HoTen = ""
        self.NamSinh = 0
        self.BangCap = ""
        self.HeSoLuong = 0.0
        self.KinhNghiem = 0
    def nhaptt(self):
        self.HoTen = input("Nhập Họ tên: ")
        self.NamSinh = int(input("Năm sinh: "))
        self.BangCap = input("Bằng cấp?")
        self.HeSoLuong = float(input("Hệ số lương: "))
        self.KinhNghiem = int(input("Năm kinh nghiệm: "))
    def xuattt(self):
        print(self.HoTen)
        print(self.TinhTuoi())
        print(self.BangCap)
        print(self.HeSoLuong)
        print(self.KinhNghiem)
        print(self.TinhLuong())
    def TinhTuoi(self):
        return datetime.now().year - self.NamSinh
    def TinhLuong(self):
        luong = (self.HeSoLuong * 1350000) + (self.KinhNghiem * 100000)
        return luong
obj = employee()
obj.nhaptt()
obj.xuattt()
```
